```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

# ËÆæÁΩÆ‰∏≠ÊñáÂ≠ó‰ΩìÔºàMicrosoft YaHei ÊòØ‰∏ÄÁßçÂ∏∏ËßÅÁöÑ‰∏≠ÊñáÂ≠ó‰ΩìÔºâ
matplotlib.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # ‰ΩøÁî® Microsoft YaHei Â≠ó‰Ωì
matplotlib.rcParams['axes.unicode_minus'] = False  # Ëß£ÂÜ≥Ë¥üÂè∑ÊòæÁ§∫ÈóÆÈ¢ò

# ÂÆö‰πâÊï∞Âàó b_n
def b_n(n):
    return 1 / n

# ÂÆö‰πâ n ÁöÑËåÉÂõ¥
n_values = np.arange(1, 41)  # ‰ªé 1 Âà∞ 40
b_values = b_n(n_values)

# ËÆ°ÁÆóÂâç10È°πÁöÑÊúÄÂ§ßÂÄºÂíåÊúÄÂ∞èÂÄº
max_value_before_10 = 1
min_value_before_10 = 1 / 10

# ËÆ°ÁÆóÁ¨¨10È°π
n_10 = 10
b_10 = b_n(n_10)

# ÂàõÂª∫ÂõæÂΩ¢ÂíåËΩ¥
plt.figure(figsize=(10, 6))
plt.plot(n_values, b_values, marker='o', linestyle='-', color='b', label='$b_n = \\frac{1}{n}$')

# Ê†áËÆ∞‰∏ä‰∏ãÁïåÁ∫ø
plt.axhline(y=1, color='r', linestyle='--', label='Ââç10È°πÁöÑ‰∏äÁïå')
plt.axhline(y=1 / 10, color='g', linestyle='--', label='Ââç10È°πÁöÑ‰∏ãÁïå')
plt.axhline(y=0, color='k', linestyle='-', label='10È°πÂêéÁöÑ‰∏ãÁïå')
plt.axhline(y=1 / 10, color='m', linestyle='--', label='10È°πÂêéÁöÑ‰∏äÁïå')

# Á™ÅÂá∫ÊòæÁ§∫Á¨¨10È°π
plt.scatter(n_10, b_10, color='r', zorder=5)
plt.text(n_10, b_10, f' ({n_10}, {b_10:.2f})', color='r', fontsize=12, ha='right')

# Ê∑ªÂä†Ê†áÈ¢òÂíåÊ†áÁ≠æ
plt.title('Êï∞Âàó $b_n = \\frac{1}{n}$')
plt.xlabel('$n$')
plt.ylabel('$b_n$')
plt.legend()

# ËÆæÁΩÆÂõæÁöÑËåÉÂõ¥
plt.xlim(0, 41)
plt.ylim(-0.1, 1.1)

# ÊòæÁ§∫ÁΩëÊ†º
plt.grid(True)

# ÊòæÁ§∫ÂõæÂΩ¢
plt.show()
```
